##lact

```python
newlist = list()
#Iterate over mylist items
for item in Lactation:
 #split the element string into a list of words
 itemWords = str(item)
 itemWords = itemWords.replace('[','')
 itemWords = itemWords.replace(']','')
 itemWords = itemWords.replace('*','')
 itemWords = itemWords.replace("'", '')
 itemWords = itemWords.replace("y", '')
 itemWords = itemWords.replace(",", '')
 
 itemWords = itemWords.split()

 #extend newlist to include all itemWords
 newlist.append(itemWords)
Lact = newlist
print(Lact)
```

```python
df = pd.DataFrame (Lact,columns=['Lower age','Upper age','Vitamin A (μg/d)','Vitamin A (μg/d)w','Vitamin C (mg/d)','Vitamin D (μg/d)','Vitamin E (mg/d)','Vitamin K(μg/d)','Thiamin (mg/d)','Riboflavin (mg/d)','Niacin (mg/d)','Vitamin B6 (mg/d)',
'Folate(μg/d)','Vitamin B12 (μg/d)','Pantothenic Acid (mg/d)','Biotin (μg/d)','Choline (mg/d)'])
df['Vitamin A (μg/d)'] = df['Vitamin A (μg/d)'] + df['Vitamin A (μg/d)w']
df = df.drop('Vitamin A (μg/d)w', 1)
df
```

```python

df =df.astype('float')
vita_lact = df
vita_lact
```
